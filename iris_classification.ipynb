# Data Preprocessing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import joblib

# Load dataset
df = pd.read_csv('iris.csv')

# Drop 'Id' column, keep only feature columns
X = df.drop(['Id', 'Species'], axis=1)  # Keep only SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm
y = df['Species']

# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

# Train Machine Learning Models

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
print("Random Forest model trained.")

# Train SVM
svm_model = SVC(probability=True, random_state=42)
svm_model.fit(X_train, y_train)
```

# Evaluate Both Models

```python
joblib.dump(rf_model, 'iris_random_forest.pkl')
joblib.dump(svm_model, 'iris_svm.pkl')
joblib.dump(le, 'label_encoder.pkl')
```

```python
import pandas as pd
import joblib

# Load models and label encoder
rf_loaded = joblib.load('iris_random_forest.pkl')
svm_loaded = joblib.load('iris_svm.pkl')
le_loaded = joblib.load('label_encoder.pkl')

# Sample input - use the exact same feature column names
sample = pd.DataFrame([[5.9, 3.0, 5.1, 1.8]], 
                      columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

# Predict
rf_pred = rf_loaded.predict(sample)
svm_pred = svm_loaded.predict(sample)

# Decode predictions back to original labels
print("Random Forest predicts:", le_loaded.inverse_transform(rf_pred)[0])
print("SVM predicts:", le_loaded.inverse_transform(svm_pred)[0])
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predict on test set
y_pred_rf = rf_model.predict(X_test)
y_pred_svm = svm_model.predict(X_test)

# Accuracy
print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred_rf):.2f}")
print(f"SVM Accuracy: {accuracy_score(y_test, y_pred_svm):.2f}")

# Classification Reports
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf, target_names=le.classes_))

print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm, target_names=le.classes_))

# Confusion Matrix Plots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

cm_rf = confusion_matrix(y_test, y_pred_rf)
cm_svm = confusion_matrix(y_test, y_pred_svm)

sns.heatmap(cm_rf, annot=True, fmt='d', ax=axes[0], xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
axes[0].set_title('Random Forest Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

sns.heatmap(cm_svm, annot=True, fmt='d', ax=axes[1], xticklabels=le.classes_, yticklabels=le.classes_, cmap='Greens')
axes[1].set_title('SVM Confusion Matrix')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

plt.show()
```
